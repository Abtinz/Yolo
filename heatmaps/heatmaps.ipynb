```python
import cv2

from ultralytics import YOLO, solutions
```

```python
model = YOLO("yolov8n.pt")
```

```python
VIDEO_URL = "./people-detection.mp4"
cap = cv2.VideoCapture(VIDEO_URL)
assert cap.isOpened(), f"error occurred in the process of reading {VIDEO_URL}"
width, height, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
print(f"{VIDEO_URL} information:", width, height, fps)
```

```python
RESULT_VIDEO_URL = "people-heatmap.mp4"
video_writer = cv2.VideoWriter(RESULT_VIDEO_URL, cv2.VideoWriter_fourcc(*"mp4v"), fps, (width, height))
```

```python
#here we initialize the Heatmap class that will illustrate the heatmaps of detected objects in real-time video stream based on their tracks
heatmap_illustrator = solutions.Heatmap(
    colormap=cv2.COLORMAP_PARULA,
    view_img=True,
    shape="circle",
    names=model.names,
)

#COLORMAP_PARULA is the common color of heatmaps illustrations 
```

```python
while cap.isOpened():
    success, frame = cap.read()
    if not success: #now the end of our video capturing is here ....
        break

    #this sector will track our objects in the given scene of our video
    tracks = model.track(frame, persist=True, show=False)

    #now lets map our tracks and the heatmap together
    frame = heatmap_illustrator.generate_heatmap(frame, tracks)
    video_writer.write(frame)

print("video processing has been successfully completed. please search the video capture for result analyzing")
cap.release()
video_writer.release()
print(f"the result is saved in the directory of: ./{RESULT_VIDEO_URL}")
cv2.destroyAllWindows()
```
